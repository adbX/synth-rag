```python
from kreuzberg import extract_file, ExtractionConfig
import numpy as np
from pathlib import Path
import json
output_dir = Path.cwd()/Path("../../outputs").resolve()
pdf_dir = Path.cwd()/Path("../../pdfs").resolve()

file_path = pdf_dir/"2012.09302.pdf"
```

```python
# Enable chunking with default chunk size and overlap
result = await extract_file(file_path, config=ExtractionConfig(chunk_content=True))
# Access the full content
full_text = result.content
```

```python
# Access individual chunks
for i, chunk in enumerate(result.chunks):
    print(f"Chunk {i+1}, length: {len(chunk)} characters")
    print(f"Preview: {chunk[:100]}...\n")
    break
```

```python
# save chunks to json
chunks_data = {
    "file": str(file_path.name),
    "total_chunks": len(result.chunks),
    "chunks": [
        {
            "chunk_id": i,
            "content": chunk,
            "length": len(chunk)
        }
        for i, chunk in enumerate(result.chunks)
    ]
}

output_file = output_dir / f"{file_path.stem}_chunks.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(chunks_data, f, indent=2, ensure_ascii=False)

print(f"Saved {len(result.chunks)} chunks to {output_file}")
```
